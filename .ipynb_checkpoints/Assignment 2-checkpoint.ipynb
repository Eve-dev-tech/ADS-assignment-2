```python
1. Show a breakdown of distance from home by job role and attrition
```

```python
# Imports needed to complete this assignment
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# how  to create a dataframe 
IBM  = {"Attrition": ["Yes", "No", "Yes"], 
            "DistanceFromHome": [1, 8, 2], 
            "JobRole": ["Sales Executive", "Research Scientist", "Laboratory Technician"]}
 
# print dataframe. 

df = pd.DataFrame(data=IBM)
```

```python
IBM 
```

```python
fig, ax = plt.subplots()
slices = [7,2,2,]
teams = ['Attrition','DistanceFromHome','JobRole']
cols = ['c','m','r']

plt.pie(slices,
        labels=teams,
        colors=cols,
        startangle=90,
        shadow= True,
        
        autopct='%1.1f%%')

plt.title('a breakdown of distance from home by job role and attrition.')
plt.show()
```
